```python
from collections import defaultdict

book = defaultdict(list)
with open('goodreads_books.json', 'r') as f:
    for line in f:
        metadata = json.loads(line.strip())
        book['title'].append(metadata['title'])
        book['description'].append(metadata['description'])
```

## length of tokens

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained(
                '/data1/home/xingmei/GRE/e5-base-v2',
                use_fast=True,
                revision='main',
                trust_remote_code=True,
            )
```

```python
len_of_title = 0
len_of_description = 0
cnt = 0

for title, description in zip(book['title'], book['description']):
    len_of_title += len(tokenizer(title)['input_ids'])
    len_of_description += len(tokenizer(description)['input_ids'])
    cnt += 1
```

```python
print(cnt, len_of_title, len_of_description, len_of_title+len_of_description, (len_of_title+len_of_description)/cnt)
```

```python
len_of_rw = 0
items = set()
with open('goodreads_reviews_dedup.json') as f:
    for line in f:
        line = json.loads(line)
        len_of_rw += len(tokenizer(line['review_text'])['input_ids'])
        items.add(line['book_id'])
print(len_of_rw,  len(items))
# 2689480686  2080190 
```
